# Plan przygotowania
* Pomijamy kolumne full_name
* Rekordy posiadające ponad połowę NaN => drop
* Rekordy posiadający diameter NaN => drop
* Uzupełnienie pozostałych wartości NaN jako średnie 

```python
import pandas as pd
pd.set_option('display.max_columns', 30)
```

```python
df = pd.read_csv('../Data/Asteroid_Updated.csv', delimiter = ',')
data_set = df.copy()

# Drop imienia
data_set.drop(columns=['name'], inplace=True)
```

```python
data_set
```

```python
data_set.info()
```

### Decyzje dotyczące wartości numerycznych:
* extent => drop (18 / 839714 wartości)
* GM => drop (14 / 839714 wartości)
* IR => drop (1 / 839714 wartość)
* G => drop (119 / 839714 wartości)

```python
data_set.drop(columns=['extent', 'GM', 'IR', 'G'], inplace=True)
data_set
```

### Drop wierszy posiadających ponad połowę NaN

```python
data_set.dropna(axis=0, thresh=13)
```

### Rekordy posiadające diameter = NaN => drop
#### Jest ich aż 702 078, więc nie będziemy uzupełniać średnią

```python
data_set = data_set[pd.notnull(data_set['diameter'])]
data_set
```

## Kolumny do sprawdzenia (condition_code, neo, pha, spec_B, spec_T, class)

```python
data_set.info()
```

### Wartości sprawdzanych kolumn

```python
for col in ('condition_code', 'neo', 'pha', 'spec_B', 'spec_T', 'class'):
    print(col + " |=>", data_set[col].value_counts())
```

### Condition_code rzutujemy wszysyko na float i zostanie 9 typów {0,1,2,3,...9}, weźmiemy pod uwagę jako numeryczna wartość

### Kolumny zawierające wartość NaN

```python
data_set.columns[data_set.isna().any()].tolist()
```

```python
# Kategoryczne dane : braki
print('spec_B NaN\'s =', data_set['spec_B'].isna().sum())
print('spec_T NaN\'s =', data_set['spec_T'].isna().sum())
```

#### Uzupełnimy NaNy najpopularniejszą wartością
spec_B            1370 non-null objects <br>
spec_T            965 non-null objects



### Numeryczne wartości uzupełnić średnią, 

```python
data_set.describe()
```

```python
data_set.fillna(data_set.mean(), inplace=True)
```

```python
# Sprawdzenie czy numeryczne wartości uzupełniono
data_set.columns[data_set.isna().any()].tolist()
```

### neo i pha rozbijemy na 2 neurony, class rozbijemy na 11 neuronów
